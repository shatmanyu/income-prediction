# GRAPHS

```python
# import matplotlib and seaborn into notebook to make graphs
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Size=formatted_df['income'].size
Size
```

```python
Greater_50K=formatted_df['income'].sum()
Greater_50K
```

```python
Less_50K=Size-Greater_50K
Less_50K
```

```python
# plot pie graph from income and peoples
distribution = np.array([Less_50K, Greater_50K])
labels = ['Below 50K', 'Above 50K']
explode = [0, 0.3]
plt.pie(x = distribution, explode=explode, labels=labels, autopct='%1.2f%%', shadow=True, startangle=45)
plt.show()
```

```python
Edu_data=formatted_df['education-num']
Edu_data.head()
```

```python
bins = np.arange(Edu_data.max()+1)
bins
```

```python
plt.hist(x = Edu_data, bins = bins, color = 'purple')
plt.xlabel('Education-Number')
plt.ylabel('People')
plt.show()
```

```python
sns.distplot(Edu_data, hist = False, color = 'm')
```
